### Joint Plots with Seaborn

```python
# generate random data
np.random.seed(42424242)
```

Generating 420 random numbers from two different gamma distributions, one with a shape parameter of 5 and the other with a shape parameter of 13. These random numbers could be used for a variety of purposes, such as statistical analysis or simulations.

```python
x = stats.gamma(5).rvs(420)
y = stats.gamma(13).rvs(420)
```

```python
# We first create a DataFrame from the x and y arrays.
# Then, we pass this DataFrame to jointplot() along with the names of the columns to plot.
# We also replace size with height to specify the size of the plot.

df = pd.DataFrame({'x': x, 'y': y})

with sns.axes_style('white'):
    sns.jointplot(data=df, x='x', y='y', kind='hex', height=10)
```
